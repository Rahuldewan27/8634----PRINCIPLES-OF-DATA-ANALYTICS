```python
# Numerical arrays and random numbers

import numpy as np

# Plot
import matplotlib.pyplot as plt
```

## Generate random integers

***



```python
rng = np.random.default_rng()
```

```python
rng.random()  
```

```python
# Generate an array of 10 numbers according to a unit Gaussian distribution.

rng.standard_normal(10)  
```

```python
# Generate an array of 5 integers uniformly over the range [0, 10).
rng.integers(low=0, high=10, size=5)
```

```python
# generate 10,000 integers over the range [0,10)

x = rng.integers(low = 0, high = 10, size = 10000, endpoint=True)

# show

x

```

```python
# Get the counts

values, counts = np.unique(x, return_counts=True)

# show

values, counts
```

```python
# Plot the histogram 

plt.bar(values, counts)
```

# Uniform Distribution 

***


```python
# generate 100 numbers uniformly over the range [0,10)

x = rng.uniform(low=0, high = 10, size =100)

#show

x
```

```python
values, counts = np.unique(x, return_counts=True)

values, counts
```

```python
plt.bar(values, counts, edgecolor='black')
```

```python
# round

x_rounded = np.round(x, 0)

values, counts = np.unique(x_rounded, return_counts=True)

# show

values, counts
```

```python
# plot

plt.bar(values, counts, edgecolor='black')
```

```python
# Generate 10000 floats uniformly over the range [0,10)

x = rng.uniform(low=0, high=10, size=10000)
```

```python
# Get the counts

counts, values = np.unique(x, return_counts=True)

#show

counts, values
```

```python
plt.hist(x, edgecolor='black')
```

```python
# Generate an array of 10000 numbers according to a unit Gaussian distribution.

x = rng.standard_normal(size=10000) 

#show

x
```
